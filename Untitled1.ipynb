```python
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
from docx import Document
import os


pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

poppler_path = r"C:\Program Files\poppler\Library\bin"  


pdf_path = "Books/G12-Ps-Civic.pdf"

images = convert_from_path(pdf_path, dpi=300, poppler_path=poppler_path)

doc = Document()

for i, img in enumerate(images):
    print(f"Processing page {i+1}...")
    text = pytesseract.image_to_string(img, lang='pus')  # 'pus' is Pashto
    cleaned = text.strip()
    if cleaned:
        doc.add_paragraph(cleaned)
    else:
        doc.add_paragraph(f"[Page {i+1} had no readable text]")

output_path = os.path.splitext(pdf_path)[0] + "_extracted.docx"
doc.save(output_path)

print(f"âœ… Extraction complete! Saved to: {output_path}")
```

```python
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
from docx import Document
import os

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

poppler_path = r"C:\Program Files\poppler\Library\bin"

pdf_path = "G12/G12-Ps-Civic.pdf"

images = convert_from_path(pdf_path, dpi=400, poppler_path=poppler_path)

doc = Document()

for i, img in enumerate(images):
    print(f"Processing page {i+1}/{len(images)}...")
    img = img.convert('L')  # grayscale for better OCR
    text = pytesseract.image_to_string(img, lang='pus')
    cleaned = text.strip()
    if cleaned:
        doc.add_paragraph(cleaned)
    else:
        doc.add_paragraph(f"[Page {i+1} had no readable text]")


output_path = os.path.splitext(pdf_path)[0] + "_extracted.docx"
doc.save(output_path)
print(f"âœ… Extraction complete! Saved to: {output_path}")
```
